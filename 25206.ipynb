```python
과목명 = input("과목명을 입력하세요: ")
학점 = float(input("학점을 입력하세요: "))
등급 = input("등급을 입력하세요 (예: A+): ")

# 등급별 평점 매핑
등급_평점 = {
    "A+": 4.5,
    "A": 4.0,
    "B+": 3.5,
    "B": 3.0,
    "C+": 2.5,
    "C": 2.0,
    "D": 1.5,
    "F": 1.0
}

전공_과목 = []

if 등급 in 등급_평점:
    전공_과목.append({
        "과목명": 과목명,
        "학점": 학점,
        "평점": 등급_평점[등급]
    })

총_학점 = sum(과목["학점"] for 과목 in 전공_과목)
총_평점_합 = sum(과목["학점"] * 과목["평점"] for 과목 in 전공_과목)

if 총_학점 > 0:
    전공평점 = 총_평점_합 / 총_학점
    print(f"전공평점: {전공평점:.6f}")
```
